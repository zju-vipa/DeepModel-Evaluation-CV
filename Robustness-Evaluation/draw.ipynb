```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_excel("ebd.xlsx")

print(data[0:15])
# 定义数据
# data = [{"name": "AlexNet",  "result": 146.242,  "acc@5":79.066, "model_size": 61.1},
# {"name": "ConvNeXt-T", "result":250.142, "acc@5":96.146, "model_size": 28.6},
# {"name": "DenseNet121", "result":245.537,"acc@5":91.972, "model_size": 8}]

new_data = data[0:15]


# 分别获取模型大小、评测结果和acc@5的数据
model_sizes = new_data["model size"] 
results = new_data["From Input(EBD)"]
accs = new_data["clean acc@5"]
# accs = new_data["Clean acc@1"]
# 计算圆的大小
sizes = [acc*5 for acc in accs]
sizes = model_sizes * 10

# import matplotlib

# font = {'family': 'serif',
#         'serif': ['Times New Roman']}
# matplotlib.rc('font', **font)

# 设置图形大小和标题
plt.figure(figsize=(10, 6))
plt.title("Model Comparison")

# 画点图，点的大小和acc@5相关，点为圆形
plt.scatter(accs, results, s=sizes,c=accs, cmap=plt.cm.RdYlBu, alpha=0.8, edgecolors='none')

for i, d in enumerate(new_data["Unnamed: 0"]):
    plt.annotate(d, xy=(accs[i], results[i]), 
                 xytext=(-5,5), textcoords='offset points', fontsize=10)

# 设置横纵坐标标签
plt.xlabel("ACC")
plt.ylabel("EBD")

# 显示图例
# plt.legend(["Model"], loc="upper left")

# 显示图形
# plt.show()
plt.savefig("ebd1.svg")
```
